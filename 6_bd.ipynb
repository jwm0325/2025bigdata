```python
import requests

url = 'http://openapi.seoul.go.kr:8088/414472704d6a776d3832534a6c4577/json/tbCycleStationInfo/1/100/'

response = requests.get(url)
```

```python
data = response.json()
data
```

```python
import pandas as pd
temp = []
for d in data['stationInfo']['row']:
    temp.append(d)

uni = pd.DataFrame(temp)
uni
```

```python
data = uni[['RENT_NM','HOLD_NUM']]
data
```

```python
data.isna()
```

```python
data.isna().sum()
```

```python
data.dropna(subset=['HOLD_NUM'])
```

```python
data.fillna(method='ffill')
```

```python
data.duplicated(['HOLD_NUM'])
```

```python
data = data.drop_duplicates(['HOLD_NUM'])
data
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data['HOLD_NUM'].value_counts().sort_index()
```

```python
data['HOLD_NUM'] = data['HOLD_NUM'].replace(45,2)
data
```
