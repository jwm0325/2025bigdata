```python
import requests
url = 'http://openapi.seoul.go.kr:8088/414472704d6a776d3832534a6c4577/json/SebcCollegeInfoKor/1/5/'
response = requests.get(url)
print(response.content)
```

```python
data = response.json()
data
```

```python
import pandas as pd
temp = []
for d in data['SebcCollegeInfoKor']['row']:
    temp.append(d)

uni = pd.DataFrame(temp)
uni
```

```python
url = 'http://openapi.seoul.go.kr:8088/414472704d6a776d3832534a6c4577/json/SeoulPublicLibraryInfo/1/5/'
r = requests.get(url)
data = r.json()
data
```

```python
l = []
for d in data['SeoulPublicLibraryInfo']['row']:
    l.append(d)

lib = pd.DataFrame(l)
lib
```

```python
import folium
dfmap = folium.Map(location=[37.4965,126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale = True)
dfmap
```

```python
for n, row in lib.iterrows():
  print(n, row['LBRRY_NAME'],row['XCNTS'],row['YDNTS'])
```

```python
from folium.plugins import MarkerCluster

mc = MarkerCluster().add_to(dfmap)
for n, row in lib.iterrows():
  mk = folium.Marker(location=[row['XCNTS'],row['YDNTS']],popup=folium.Popup(row['LBRRY_NAME'],max_width=200))
  mk.add_to(mc)
dfmap
```
