```python
import pandas as pd
df = pd.read_excel('/content/자연휴양림출력화일.xlsx')
df
```

```python
import folium
dfmap = folium.Map(location=[37.4965,126.9573],
                   zoom_start=11,
                   zoom_control=False,
                   control_scale = True)
dfmap
```

```python
for n, row in df.iterrows():
  print(n, row['휴양림명'],row['위도'],row['경도'])
```

```python
from folium.plugins import MarkerCluster

mc = MarkerCluster().add_to(dfmap)
for n, row in df.iterrows():
  mk = folium.Marker(location=[row['위도'],row['경도']],popup=folium.Popup(row['휴양림명'],max_width=200))
  mk.add_to(mc)
dfmap
```
